```python
import pandas as pd
```

```python
flight = pd.read_excel('Data_Train.xlsx')
```

```python
flight
```

```python
flight.info()
```

```python
flight.dropna(inplace=True)
```

```python
flight.info()
```

```python
flight['JourneyDay'] = pd.to_datetime(flight['Date_of_Journey'])
```

```python
flight['Day'] = pd.to_datetime(flight['JourneyDay'],format = "%d%m%Y").dt.day
```

```python
flight['Month'] = pd.to_datetime(flight['JourneyDay'],format = "%d%m%Y").dt.month
```

```python
flight.drop(['Date_of_Journey','JourneyDay'],axis='columns',inplace=True)
```

```python
flight['Total_Stops'].value_counts()
```

```python
flight['Total_Stops'].replace({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4},inplace=True)
```

```python
flight
```

```python
flight['Dep_Hour'] = pd.to_datetime(flight['Dep_Time']).dt.hour 
flight['Dep_Min'] = pd.to_datetime(flight['Dep_Time']).dt.minute 
```

```python
flight.drop(['Dep_Time'],axis = 'columns',inplace = True)
flight
```

```python
flight['Arrival_Hour'] = pd.to_datetime(flight['Arrival_Time']).dt.hour 
flight['Arrival_Min'] = pd.to_datetime(flight['Arrival_Time']).dt.minute 
flight.drop(['Arrival_Time'],axis = 'columns' , inplace = True)
flight
```

```python
flight.drop(['Duration'],axis = 'columns',inplace = True)
```

```python
flight.drop(['Route','Additional_Info'],axis = 'columns',inplace = True)
```

```python
flight
```

```python
Airline = flight['Airline']
Airline = pd.get_dummies(Airline)
```

```python

flight['Source'].value_counts()
```

```python
Source = flight['Source']
Destination = flight['Destination']
Source = pd.get_dummies(Source)
Destination = pd.get_dummies(Destination)
```

```python
final = pd.concat([flight,Airline,Source,Destination],axis=1)
final.drop(['Airline','Source','Destination'],axis = 'columns',inplace = True)
final
```

```python
target = final['Price']
final.drop(['Price'],axis = 'columns',inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final,target,test_size=0.2,random_state = 42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, final, target, cv=10)
scores.mean()
```
